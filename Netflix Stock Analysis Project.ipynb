## NETFLIX STOCK ANALYSIS PROJECT

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
netflix = pd.read_csv("Netflix.csv")
```

```python
netflix
```

```python
netflix.info()
```

```python
netflix["Date"] = pd.to_datetime(netflix["Date"])
```

```python
netflix.set_index("Date",inplace=True)
```

```python
netflix
```

### 1.Volume of Stock Traded

```python
sns.lineplot(x=netflix.index,y=netflix["Volume"],label="Volume Trade")
plt.title("Volume Stock versus Date")
```

### 2.Netflix Stock Price - high,open,close

```python
fig,ax = plt.subplots(3,1,sharex=True)
ax[0].plot(netflix.index,netflix["High"])
ax[1].plot(netflix.index,netflix["Open"])
ax[2].plot(netflix.index,netflix["Close"])

ax[0].set_ylabel("High")
ax[0].set_title("Highest Stock Price")
ax[1].set_ylabel("Open")
ax[1].set_title("Opening Stock Price")
ax[2].set_xlabel("Year")
ax[2].set_ylabel("Close")
ax[2].set_title("Closing Stock Price")
plt.show()
```

## 3.Netflix Stock Price - Day,Month,Yearwise

```python
fig, (ax1,ax2,ax3)= plt.subplots(3, figsize=(15,10))
netflix.groupby(netflix.index.day).mean().plot(y="Volume", ax=ax1, xlabel="Day",marker="o",color="r")
netflix.groupby(netflix.index.month).mean().plot(y="Volume", ax=ax2, xlabel="Month",marker="o",color="k")
netflix.groupby(netflix.index.year).mean().plot(y="Volume", ax=ax3, xlabel="Year",marker="o",color="g")
```

## 4. Top-5 Dates With HIghest Stock price

```python
Top_5_high = netflix.sort_values(ascending=False, by="High").head(5)
Top_5_high["High"]
```

## 5. Top 5 Dates with Lowest Stock Price

```python
Top_5_low = netflix.sort_values(ascending=True, by="Low").head(5)
Top_5_low["Low"]
```
